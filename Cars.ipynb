### Convert data type of  columns Invoice and MSRP into float

```python
df['Invoice'] = df['Invoice'].astype(str)
df['MSRP'] = df['MSRP'].astype(str)
df['Invoice'] = df['Invoice'].str.replace('$', '', regex=False).str.replace(',', '',regex=False).astype(float)
df['MSRP'] = df['MSRP'].str.replace('$', '', regex=False).str.replace(',', '',regex=False).astype(float)
```

```python
df.info() #df.dtypes
```

```python
df.describe()
```

### Add a new Column Price Difference

```python
df['Price Difference'] = df['MSRP'] - df['Invoice']
df
```

### Finding Null Values and then fill with the mean

```python
df.isnull().sum()
```

```python
df['Cylinders'].fillna(df['Cylinders'].mean(), inplace = True)
df.isnull().sum()
```

### Correlation Matrix

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

### Insights
- Engine size and horsepower have the strongest positive correlation with price difference. This means that cars with larger engines and more horsepower tend to have a higher MSRP than invoice price, indicating that there is more room for negotiation on these vehicles.
- MPG city and MPG highway have a negative correlation with price difference. This means that cars with better fuel economy tend to have a lower MSRP than invoice price, indicating that there is less room for negotiation on these vehicles.
- Weight, wheelbase, and length have a weak correlation with price difference. This means that these features are not as significant in determining the MSRP and invoice price of a car.

### Highest  and Lowest Profit Margins

```python
# Find the car with the highest profit margin
highest_profit_car = df[df['Price Difference'] == df['Price Difference'].max()]

# Find the car with the lowest profit margin
lowest_profit_car = df[df['Price Difference'] == df['Price Difference'].min()]

# Display the DataFrame with profit margin, average, highest, and lowest profit cars
print("\nHighest Profit Margin Car:\n", highest_profit_car)
print("\nLowest Profit Margin Car:\n", lowest_profit_car)
```

### Insights
- The Porsche GT2 2dr model boasts the highest profit margin, which amounts to $173,560.

- The Suzuki Forenza S 4 door Sedan model produced by Suzuki has the lowest profit margin, which amounts to $12,116

### Unique count of specific columns in the dataset.

```python
distinct_counts = {
    'Make': df['Make'].nunique(),
    'Model': df['Model'].nunique(),
    'Type': df['Type'].nunique(),
    'Origin': df['Origin'].nunique(),
    'EngineSize': df['EngineSize'].nunique(),
    'Cylinders': df['Cylinders'].nunique()
}
distinct_counts_df = pd.DataFrame(list(distinct_counts.items()), columns=['Features', 'Distinct Count'])
print(distinct_counts_df)
```

### Identifying the maximum wheellbase for diffenent types

```python
max_wheelbase_by_type = df.groupby('Type')['Wheelbase'].max().reset_index()
dfmax = pd.DataFrame(max_wheelbase_by_type)
dfmax
```

### Insights
- Trucks have the highest maximum wheelbase among all the types, with a maximum wheelbase of 144 inches. This suggests that trucks tend to be larger and have longer chassis, which can contribute to their stability and cargo capacity.

### Identifying the minimum wheellbase for diffenent types

```python
min_wheelbase_by_type = df.groupby('Type')['Wheelbase'].min().reset_index()
dfmin = pd.DataFrame(min_wheelbase_by_type)
dfmin
```

### Insights
- Sports cars have the shortest minimum wheelbase among all the types, with a minimum wheelbase of 89 inches. This aligns with the characteristics of sports cars, which typically have compact and agile designs to enhance their performance and handling.

### Analysis of Car Types and Origins
- I present an analysis of car data that focuses on the types of cars and their respective origins.
- The pivot table below provides a comprehensive overview of the distribution of car makes (Make) across different car types (Type) and origins (Origin).

```python
pivot_table = pd.pivot_table(df, values='Make', index='Type', columns='Origin', aggfunc='count', fill_value=0)
pivot_table
```

### Insights from Pivot Table

1. **Origin Distribution by Vehicle Type**:
   - "Hybrid" vehicles predominantly originate from Asia.
   - "SUV" and "Sedan" types have a significant presence of vehicles from Asia and the USA.
   - European origins are notable for "Sedan" and "Sports" type vehicles.

2. **Type Distribution by Origin**:
   - "Sedan" type vehicles are widely produced and have origins in all three regions.
   - "SUV" type vehicles are prevalent across regions, particularly in the USA.
   - Regional preferences are evident, with each region specializing in specific vehicle types.

3. **Regional Preferences and Diversity**:
   - Regional preferences for vehicle types are reflected, with Asia, Europe, and the USA specializing in different categories.
   - "Sedan" type vehicles exhibit diversity, being produced in significant numbers in all three regions.

### Car Make and Type Distribution
- The following pivot table provides insights into the distribution of car makes (Make) across various car types (Type).
- The values in the table represent the counts of each car make within each car type category. 
- This analysis allows us to understand how different car makes are distributed across different car types.

```python
pivot_table1 = df.pivot_table(index='Make', columns='Type', aggfunc='size', fill_value=0)
pivot_table1
```

### Insights from Pivot Table: Vehicle Type Distribution by Make

1. **Variety in Vehicle Types**: The dataset showcases a diverse array of vehicle types, with makes producing a combination of "SUV," "Sedan," "Sports," "Truck," "Wagon," and "Hybrid" vehicles.

2. **Dominant Types by Make**: "Sedan" and "SUV" are prevalent vehicle types across many makes, while some specialize in unique types like "Sports" or "Truck." Makes like Mercedes-Benz offer a wide range of vehicle types, reflecting their versatility.

3. **Unique Offerings**: Certain makes stand out with distinctive offerings not widely represented by others. For instance, Porsche specializes in "Sports" cars, and Acura, Honda, and Toyota produce "Hybrid" vehicles, highlighting unique market positions.
